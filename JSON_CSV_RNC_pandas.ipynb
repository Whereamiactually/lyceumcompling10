# JSON

JSON (англ. JavaScript Object Notation) — текстовый формат обмена данными, основанный на JavaScript (при этом формат независим от JS и может использоваться в любом языке программирования). Для этих же целей ещё используются форматы XML и YAML, но о них мы говорить не будем.

В качестве значений в JSON могут быть использованы:

* JSON-объект (неупорядоченное множество пар «ключ : значение», "словарь"),
* Массив,
* Число (целое или вещественное),
* Булевые значения True, False, а также специальный тип Null/None ("ничего"),
* Строка.



```python
import json
```

JSON-объект по сути являет собой словарь - неупорядоченное множество пар "ключ : значение". Его можно записать с помощью строки, по структуре представляющей собой словарь, а можно и с помощью обычного словаря.

Ключ должен быть строкой (и обязательно в двойных кавычках!), даже если это число. Строки в значниях тоже должны быть в двойных кавычках!

```python
json_str = """{
  "firstName": "Григорий",
  "lastName": "Печорин",
  "hobbies": ["стреляться", "противоречить", "ухаживать за женщинами"],
  "age": 25
}""" # строка

json_json = json.loads(json_str) # делает из объекта JSON словарь, то есть декодирует
print(json_json)
print(json_json["age"])
print(type(json_json))
```

```python
json_dict = {
  "FirstName": "Grigory",
  "LastName": "Печорин",
  "Hobbies": ["стреляться", "противоречить", "ухаживать за женщинами"],
  "Age": 25,
  "Children": [
        {
            "FirstName": "Алиса",
            "Age": 6
        },
        {
            "FirstName": "Stanislav",
            "Age": 8
        }
    ]
} # словарь

json_json = json.dumps(json_dict, indent = 1) # делает из словаря объект JSON, то есть кодирует
# indent задаёт уровень отступов результирующей строки
print(json_json) # так как это строка, то нельзя вытащить значение по ключу, как мы сделали это со словарём
# кириллические символы показываются в кодировке UTF8
print(type(json_json))
```

Когда вы конвертируете из Python'a в JSON, объекты Python'a конвертируются в их эквиваленты JavaScript.

NB: если

| Python | JSON |
|----------|----------|
| Dictionary (словарь)	| Object (объект) |
| List	(список) | Array (массив) |
| Tuple	(кортеж) | Array (массив) |
| String (строка) |	String (строка) |
| Integer (целое число) |	Number (число) |
| Float	(вещественное число) | Number (число) |
| True	| True |
| False	| False |
| None	| Null |

Чаще всего мы будем работать именно с объектами.

```python
print(json.dumps({"Name": "John", "Age": 99}))
print(json.dumps(["vowels", "consonants"]))
print(json.dumps(("vowels", "consonants")))
print(json.dumps("Colorless green ideas sleep furiously."))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
print(json.dumps(json_dict, indent = 4, separators = (". ", " = "))) # можно поменять разделители между парами и между ключами и их значениями
```

```python
print(json.dumps(json_dict, indent = 4, sort_keys = True)) # можно отсортировать по ключам (по убыванию)
```

Теперь посмотрим на (почти) реальные примеры.
Мы будем использовать [JSONPlaceholder](https://jsonplaceholder.typicode.com/), который содержит данные JSON.

```python
import requests
```

```python
response = requests.get("https://jsonplaceholder.typicode.com/todos")
todos = json.loads(response.text) # возвращает содержание ответа
```

```python
print(type(todos))
print(todos[:10])
print(len(todos))
```

В каждом объекте пользователь (**userID**) соотносится с одним из двадцати заданий (**id**, **title**), относительно которого известно, выполнил пользователь его или нет (**completed**).

Давайте напишем код, который определяет пользователей, выполнивших максимальное количество заданий.

```python
todos_by_user = {}

for todo in todos:
    if todo["completed"] and todo["userId"] in todos_by_user:
      todos_by_user[todo["userId"]] += 1
    elif todo["completed"] and todo["userId"] not in todos_by_user:
      todos_by_user[todo["userId"]] = 1

top_users = sorted(todos_by_user.items(), # сортирует по убыванию по количеству (x[1]) решённых задач
                   key = lambda x: x[1], reverse = True)

print("Пользователи по убыванию:", top_users)
max_complete = top_users[0][1] # находим нибольшее количество завершённых заданий

print("Наибольшее количество завершённых заданий:", max_complete)

users = []
for user, num_complete in top_users:
    if num_complete == max_complete: # собираем пользователей, который выполнили максимум заданий
      users.append(str(user))

max_users = " и ".join(users)
print("ID пользователей, которые решили больше всех заданий:", max_users)
```

Это ничем не отличается от того, что мы делаем с объектами Python'a. Конечно, ведь мы просто получили данные посредством JSON, декодировали их, а потом использовали их в наших вычислениях.

Мы можем также создать собственный JSON-объект.

Давайте запишем все выполненные тексты для каждого пользователя, выполнившего максимальное количество заданий.

```python
def keep(todo): # создаём собственную функцию
# она берёт на вход  один из объектов в TODOs
# и выдаёт, является ли он выполненным и содержится ли его пользователь в списке пользователей, выполнивших максимум
    is_complete = todo["completed"] #
    max_count = str(todo["userId"]) in users
    return is_complete and max_count

with open("filtered_data_file.json", "w") as data_file: # записываем информацию в файл
    filtered_todos = list(filter(keep, todos)) # отфильтровываем TODOs, которые содержат информацию, соответствующую нашему заданию
    json.dump(filtered_todos, data_file, indent = 2)
```

Теперь посмотрим на совсем реальный пример работы с данными.

```python
colnames = ['Left context', 'Center', 'Punct', 'Right context', 'Title', 'Full context' ]
data = pd.read_csv("RNC_witless.csv", sep = ';', usecols = colnames)
data.head()
```

Например, мы хоти понять, насколько часто слово "умалишённый" встречается без последующего существительного. Какой у нас может быть алгоритм?

```python
new_data = data[['Center', 'Right context']]
new_data.head()
```

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
from razdel import tokenize
```

```python
right_pos = [] # сюда будем класть часть речи правого контекста
witless_morph = [] # сюда будем класть разбор исследуемого слова
right_morph = [] # сюда будем класть разбор правого контекста
for index, row in new_data.iterrows(): # проходимся по каждой строке в таблице
    main_case = str(morph.parse(row['Center'])[0].tag.case) # вытаскиваем падеж из анализа слова
    witless_morph.append(main_case) # собираем их в список
    try: # пробуем определить морф. параметры слова из правого контекста, если оно есть
      context_word = list(list(tokenize(row['Right context']))[0])[2] #
      context_pos = str(morph.parse(context_word)[0].tag.POS) #
      context_case = str(morph.parse(row['Center'])[0].tag.case)
    except: # если его нет, то ошибку не выдаёт, а приписывает определённое значение
      context_pos = "No context."
      context_case = "No context."
    right_pos.append(context_pos) # собираем в список все части речи
    right_morph.append(context_case) # собираем в список все падежи
```

```python
new_data['Main case'] = witless_morph # добавляем новые колоночки
new_data['Right POS'] = right_pos
new_data['Right case'] = right_morph
```

```python
new_data.head()
```

Теперь создадим столбец "Result", в который будет вноситься "with noun" ("с существительным"), если в правом контексте содержится существительное и падеж исследуемого слова равен падежу этого существительного.

```python
import numpy as np
```

```python
new_data['Result'] = np.where((new_data['Right POS'] == 'NOUN') & (new_data['Main case'] == new_data['Right case']), "with noun", np.nan)
new_data.head() # добавляем новую колоночку, в которой будет результат сравнения
```

Теперь отфильтруем и посчитаем.

```python
filtered_data = new_data.loc[new_data.Result == 'with noun']
filtered_data # отфильтровываем по тому, связано ли наше прилагательное с существительным
```

```python
len_all = len(new_data) # считаем длины таблиц
len_with_noun = len(filtered_data)
print(f"Только {len_with_noun} вхождений исследуемоего слова (из {len_all}) встречаются с существительными.")
```
